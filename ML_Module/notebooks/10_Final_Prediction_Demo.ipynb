# 10. Final Prediction Demo
A live simulation cell for external review / viva showcasing multi-model consensus.

```python
import joblib
import pandas as pd

try:
    rf_model = joblib.load("../models/random_forest.joblib")
    xgb_model = joblib.load("../models/xgboost_model.joblib")
    
    # Mock input mapping to: Age, Salary, Experience, Weekly_Hours, Tasks_Completed, Department_Encoded, Gender_Encoded
    mock_input = pd.DataFrame([[0.5, -0.2, 1.2, 1.8, 2.0, 0, 1]], 
        columns=['Age', 'Salary', 'Experience', 'Weekly_Hours', 'Tasks_Completed', 'Department_Encoded', 'Gender_Encoded'])
    
    print("--- LIVE MULTI-MODEL PREDICTION DEMO ---")
    print("Simulated Employee Features:\n", mock_input)
    
    rf_pred = rf_model.predict(mock_input)[0]
    xgb_pred = xgb_model.predict(mock_input)[0]
    
    print("\nRandom Forest Predicts:", 'High Performer' if rf_pred == 1 else 'Standard Performer')
    print("XGBoost Predicts:      ", 'High Performer' if xgb_pred == 1 else 'Standard Performer')
    
    if rf_pred == xgb_pred:
        print("\nConsensus Reached! The models agree on the classification.")
    else:
        print("\nModels Disagree. This highlights the probabilistic nature of realistic ML applications.")
    
except Exception as e:
    print("Models missing.", e)
```
